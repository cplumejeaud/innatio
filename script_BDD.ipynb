1. Importation des données

````python
#Chargement des codes EPCI et départements
import pandas as pd
cols= ['EPCI', 'DEP']
EPCI= pd.read_excel(r"C:\Users\julia\OneDrive\Documents\Python_mineure_M2\HTML\projet\EPCI.xlsx", usecols=cols)
EPCI.head()
````

````python
EPCI= EPCI.rename(columns={"DEP":"num_dep"})
EPCI.head()
````

````python
dpt_region = pd.read_csv(r"C:\Users\julia\OneDrive\Documents\Python_mineure_M2\HTML\projet\departements-region.csv",
    sep=",", encoding='UTF-8')

dpt_region.head()
````

````python
region_EPCI= EPCI.merge(dpt_region, left_on ="num_dep", right_on="num_dep", how="left")
region_EPCI["EPCI"] = region_EPCI["EPCI"].astype(str)
region_EPCI.head()
````

````python
#Supprimer les doublons dans region_EPCI
region_EPCI = region_EPCI.drop_duplicates(subset="EPCI")
````

## Attention : il y a des EPCI à cheval sur deux régions. Donc là vous les supprimez ! 

````python
INAT_NAT = pd.read_csv(
    r"C:\Users\julia\OneDrive\Documents\Python_mineure_M2\HTML\projet\_INAT_NAT_EPCI_2025-09-15.csv",
    sep=";",
    encoding='latin1',
    dtype={"unit": str}  
)
````

````python
NAT_EPCI = pd.read_csv(
    r"C:\Users\julia\OneDrive\Documents\Python_mineure_M2\HTML\projet\_NAT_EPCI_2025-09-15.csv",
    sep=";",
    encoding='latin1',
    dtype={"unit": str}  # remplace "EPCI" par le nom exact de ta colonne
)
````

````python
#renomer la colonne "unit" de Inat_Nat en "EPCI"
INAT_NAT= INAT_NAT.rename(columns={"unit":"EPCI"})
INAT_NAT["EPCI"] = INAT_NAT["EPCI"].astype(str)
INAT_NAT.head()
````

````python
#renomer la colonne "unit" de NAT_EPCI en "EPCI"
NAT_EPCI= NAT_EPCI.rename(columns={"unit":"EPCI"})
NAT_EPCI["EPCI"] = NAT_EPCI["EPCI"].astype(str)
NAT_EPCI.head()
````

````python
#Réaliser une jointure pour ajouter les départements et régions aux données INAT_NAT sur la variable EPCI
INAT_NAT_REGION= INAT_NAT.merge(region_EPCI, left_on ="EPCI", right_on="EPCI", how="left")
INAT_NAT_REGION.head()
````

````python
#enregistrer les nouvelles bases de données avec les régions et départements
INAT_NAT_REGION.to_csv(r"C:\Users\julia\OneDrive\Documents\Python_mineure_M2\HTML\projet\INAT_NAT_EPCI_REGION.csv", index=False, sep=';',na_rep="NA" )
````

````python
NAT_EPCI_REGION= NAT_EPCI.merge(region_EPCI, left_on ="EPCI", right_on="EPCI", how="left")
NAT_EPCI.head()
````

````python
#enrichir NAT_EPCI_REGION  avec les informations de département et région
NAT_EPCI_REGION.to_csv(r"C:\Users\julia\OneDrive\Documents\Python_mineure_M2\HTML\projet\NAT_EPCI_REGION.csv", index=False, sep=';', na_rep="NA")
````
